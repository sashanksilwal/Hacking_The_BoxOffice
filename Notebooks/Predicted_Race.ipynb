```python
mcgill_actors_df = pd.read_csv("mcgill_actors.csv")
mcgill_actors_df["nconst"] = mcgill_actors_df.IMDB_ACTOR_HOMEPAGE.apply(lambda x: x.split("/")[-2])
mcgill_actors_df
```

```python
len(predicted_race_df[predicted_race_df.nconst.isin(mcgill_actors_df.nconst)].nconst.unique())
```

```python
merged_with_mcgill = pd.merge(predicted_race_df, mcgill_actors_df, left_on='nconst', right_on='nconst', how='left')
merged_with_mcgill = merged_with_mcgill.drop_duplicates(subset=['nconst'], keep='last')
```

```python
clean_merged_with_mcgill = merged_with_mcgill[["first_name","last_name","NEW_ACTOR_RACE", "race","nconst" ]]
```

```python
clean_merged_with_mcgill_NA = clean_merged_with_mcgill.copy()
clean_merged_with_mcgill_NA["NEW_ACTOR_RACE"] = clean_merged_with_mcgill["NEW_ACTOR_RACE"].fillna(clean_merged_with_mcgill['race'].apply(lambda x:  x[-1]  if x[-1] in ["W","B", "A","P", "I"] else x ))#x[0:2]  ))

```

```python
clean_merged_with_mcgill_NA.reset_index(drop=True)
```

```python
clean_merged_with_mcgill_NA["NEW_ACTOR_RACE"] = clean_merged_with_mcgill_NA["NEW_ACTOR_RACE"].replace(['s', 'n', 'I', 'i', 'W', 'B','A'], ['a',"NL+O", 'i', 'a','w','b','a'])
```
