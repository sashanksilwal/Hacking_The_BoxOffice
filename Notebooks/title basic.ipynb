```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = '/content/drive/MyDrive/Year 4 (2022 - 2023)/Fall 2022/Computational Social Science/Project/title_basic_data.tsv'
```

```python
df = pd.read_csv(url, delimiter='\t')
df.head()
```

```python
df.to_csv('title_basic_data.csv', index=False)
```

```python
df2 = pd.read_csv('title_basic_data.csv')
df2.head()
```

```python
df.tconst.unique().shape
```

```python
df.drop(df[df['startYear'] == '\\N'].index, inplace = True)
df.drop(df[df['runtimeMinutes'] == '\\N'].index, inplace = True)
df.head()
```

```python
df.shape
```

```python
df.startYear = df.startYear.astype('int64')
df.runtimeMinutes = df.runtimeMinutes.astype('int64')
```

```python
df = df[df.titleType == 'movie']
df.head()
```

```python
# df.startYear.describe()
bins = np.arange(2010, 2021)

plt.figure(figsize=(8, 6))
plt.hist(df.startYear, bins=bins, color='lightsalmon', edgecolor='white')
plt.xticks(bins)
plt.ylabel('Number of movies')
plt.xlabel('Year')
plt.title("Distribution of movies from 1900 - 2020", fontdict={'fontsize': 14})
plt.show()
```

```python
plt.scatter(df.startYear, df.runtimeMinutes)
```

```python
bins = np.arange(40, 320, 20)

plt.figure(figsize=(8, 6))
plt.hist(df.runtimeMinutes, bins=bins, color='paleturquoise', edgecolor='white')
plt.xticks(bins)
plt.ylabel('Number of movies')
plt.xlabel('Runtime (minutes)')
plt.title("Distribution of movies with runtime between 40 and 300 minutes", fontdict={'fontsize': 14})
plt.show()
```
